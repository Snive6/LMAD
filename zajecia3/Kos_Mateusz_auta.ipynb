Proszę przygotować rozwiązanie (program, funkcję, polecenie) w języku Python, które pozwoli na segmentację modeli samochodów, za pomocą poznanych metod grupowania.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN

from plotnine import *

from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('auta.csv', index_col=0)
```

```python
auta = df.copy()
auta.head(10)
```

```python
z = StandardScaler()
auta[auta.columns] = z.fit_transform(df[df.columns])
```

```python
ks = range(2, 20)

sylwetka_km = []
sylwetka_gm = []
sylwetka_db = []

for k in ks:
    km = KMeans(n_clusters=k).fit(auta)
    gm = GaussianMixture(n_components=k).fit(auta)
    db = DBSCAN(eps=0.5 + k*0.1, min_samples=5).fit(auta)
    df['klastry_km'] = km.labels_
    df['klastry_gm'] = gm.predict(auta)
    df['klastry_db'] = db.labels_
    sylwetka_km.append(round(silhouette_score(auta, km.labels_), 4))
    sylwetka_gm.append(round(silhouette_score(auta, gm.predict(auta)), 4))
    sylwetka_db.append(round(silhouette_score(auta, db.labels_), 4))
```

```python
epsilon = [round(0.5 + k * 0.1, 2) for k in ks]
```

```python
sylwetki_km_df = pd.DataFrame({'K': ks,
                           'Sylwetka': sylwetka_km})
sylwetki_gm_df = pd.DataFrame({'K': ks,
                           'Sylwetka': sylwetka_gm})
sylwetki_db_df = pd.DataFrame({'Epsilon': epsilon,
                           'Sylwetka': sylwetka_db})
```

```python
for k, v in {"KM": sylwetki_km_df, "GM": sylwetki_gm_df}.items():
    print((ggplot(v, aes(x = 'K', y='Sylwetka'))+
     geom_point() +
     scale_x_continuous(breaks = ks) +
     geom_line() +
     theme_minimal() +
     labs(title= f'Sylwetki dla kolejnych K - {k}')))
```

```python
print((ggplot(sylwetki_db_df, aes(x = 'Epsilon', y='Sylwetka'))+
     geom_point() +
     scale_x_continuous(breaks = epsilon) +
     geom_line() +
     theme_minimal() +
     labs(title= f'Sylwetki dla kolejnych wartości Epsilon')))
```

Według KMeans, GaussianMixture oraz DBSCAN optymalną liczbą grup dla tych danych będą dwie. Największą (najlepszą) wartość sylwetki dla tych danych okazało się DBSCAN.
<br> sylwetka DBSCAN = 0.6572
<br> sylwetka KMEANS = 0.6241
<br> sylwetka DBSCAN = 0.4342

```python
km = KMeans(n_clusters=2).fit(auta)
gm = GaussianMixture(n_components=2).fit(auta)
db = DBSCAN(eps=1.9, min_samples=5).fit(auta)

df['klastry_km'] = km.labels_
df['klastry_gm'] = gm.predict(auta)
df['klastry_db'] = db.labels_
```

```python
for zm in auta.columns:
    print((ggplot(df, aes(x='klastry_km', y=zm, fill = 'factor(klastry_km)'))+
 geom_boxplot(show_legend=False) +
 theme_minimal() +
 labs(x = 'Klastry KM', title = zm)))
```

```python
for zm in auta.columns:
    print((ggplot(df, aes(x='klastry_gm', y=zm, fill = 'factor(klastry_gm)'))+
 geom_boxplot(show_legend=False) +
 theme_minimal() +
 labs(x = 'Klastry GM', title = zm)))
```

```python
for zm in auta.columns:
    print((ggplot(df, aes(x='klastry_db', y=zm, fill = 'factor(klastry_db)'))+
 geom_boxplot(show_legend=False) +
 theme_minimal() +
 labs(x = 'Klastry DBSCAN', title = zm)))
```

### Komentarz dla dwóch algorytmów, czyli KMeans oraz DBSCAN (w obu przypadkach wyniki się pokrywają):

grupa 0 charakteryzuje sie wysokim przebiegiem, niska cena, niska pojemnoscia i niska liczba KM <br>
grupa 1 (-1) charakteryzuje sie niskim przebiegiem, wysoka cena, wysoka pojemnoscia oraz wysoka liczba KM
